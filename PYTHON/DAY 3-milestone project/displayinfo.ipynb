```python
def user_choice():
    choice ='wrong'
    accept_range = range(1,11)
    within_range = False
    while choice.isdigit() == False or within_range == False:
        choice = input('enter a num 1-10:')

        if choice.isdigit() == False:
            print('sorry that is not a digit')

        if choice.isdigit() == True:
            if int(choice) in accept_range:
                within_range = True
            else :
                 print('sorry that is a digit with out of range')
                 within_range = False

            
    return int(choice)

user_choice()
```

```python
def position_choice():
    choice='wrong'

    while choice not in ['0','1','2']:

        choice = input('pick a position (0,1,2) :')

        if choice not in ['0','1','2']:
            print ( 'sorry invalid choice' )


    return int(choice)
```

```python
position_choice()
```

```python
def replacement_choice(game_list,position):

    user_placement = input(' type a string to place at position: ')

    game_list[position] = user_placement 

    return game_list
```

```python
 replacement_choice(game_list,position)
```

```python
 replacement_choice(game_list,2)
```

```python
def gameon_choice():
    choice='wrong'

    while choice not in ['Y','N']:

        choice = input('keep playing? (Y or N) ')

        if choice not in ['Y','N']:
            print ( 'sorry invalid choice please choose  (Y or N) ' )


    if choice == 'Y':
       return True
    else :
           return False


```

```python
gameon_choice()
```

```python
game_on = True
game_list=[0,1,2]

while game_on:

    display_game(game_list)
    position= position_choice()
    game_list=replacement_choice(game_list,position)
    display_game(game_list)
    game_on=gameon_choice()
    
```
