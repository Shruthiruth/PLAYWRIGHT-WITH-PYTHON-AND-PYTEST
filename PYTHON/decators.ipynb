```python
def new_decorator(original_func):
    def wrap_Func():
        print('some extra code before orifinal function')

        original_func()
        print('after original function some extra code')

    return wrap_Func
```

```python
def func_needs_decorator():
    print('i want a decorator')
    
```

```python
func_needs_decorator()
```

```python
some_decorated_func=new_decorator(func_needs_decorator)
```

```python
some_decorated_func()
```
